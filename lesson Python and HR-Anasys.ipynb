# Функции

```python
nums = [1,2,3,4,5]
nums
```

```python
def multy(nums):
    new = []
    for i in nums:
        new.append(i*2)
    return new
```

```python
multy([6,7,8,9])
```

```python
def multy_k(nums,k):
    new = []
    for i in nums:
        new.append(i*k)
    return new
```

```python
multy_k([1,2,3,4,5],6)
```

# Классы

```python
specialist = [('Jon', 'DE', 150), ('Carl', 'DA', 120), ('Gleb', 'DS', 300)]
```

```python
class Specialict:
    def __init__(self, name, speciality, salary):
        self.name = name
        self.speciality = speciality
        self.salary = salary
        
    def tell_about_yourself(self):
        return f'Hi, my name is {self.name}. I\'m a {self.speciality}. My salary is {self.salary}'
```

```python
Jon = Specialict('Jon', 'DE', 150)
```

```python
Jon
```

```python
Jon.salary
```

```python
Carl = Specialict('Carl', 'DA',120)
```

```python
Carl.speciality
```

```python
Carl.tell_about_yourself()
```

# Наследование классов

```python
class DataScientist(Specialict):
    
    def __init__(self, name, salary):
        super().__init__(name,'DS', salary)
        
    def tell_about_yourself(self):
        return f'Hi, my name is {self.name}. I\'m a {self.speciality}. My salary is {self.salary}k'
    
    def ml(self, data):
        return 'profit for company (and myself)'
```

```python
Gleb = DataScientist('Gleb', 300)
```

```python
Gleb.tell_about_yourself()
```

```python
Gleb.ml([1,2,34567, -1])
```
